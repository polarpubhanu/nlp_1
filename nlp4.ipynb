```python
import pandas as pd
import re

data_out = pd.DataFrame({
    'twitts': ['hello123world', 'python3 is great', 'data2024science']
})

data_out['twitts_clean'] = data_out['twitts'].apply(
    lambda x: re.sub(r'\d+', '', x)
)

print(data_out)
```

```python
import re

data_out['twitts'] = data_out['twitts'].apply(
    lambda x: re.sub(r'http\S+|www\S+', '', x)
)
```

```python
# remove special characters & punctuation
data_out['twitts'] = data_out['twitts'].apply(
    lambda x: re.sub(r'[^A-Za-z\s]', '', x)
)

# convert to lowercase
data_out['twitts'] = data_out['twitts'].str.lower()

# remove extra spaces
data_out['twitts'] = data_out['twitts'].str.strip()
```

```python
import nltk
nltk.download('stopwords')

from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

data_out['twitts'] = data_out['twitts'].apply(
    lambda x: ' '.join(word for word in x.split() if word not in stop_words)
)
```
